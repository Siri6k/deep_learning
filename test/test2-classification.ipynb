Classification - M√©thode de Nearest Neighbors

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier

plt.style.use('dark_background')  # üï∂Ô∏è Dark mode
```

```python
# creation des 50 points
X, y = make_blobs(n_samples=50, centers=2, n_features=2, random_state=0, cluster_std=1)

X_new = np.random.uniform(low=X.min(), high=X.max(), size=(10, 2))

plt.scatter(X[:, 0], X[:, 1], c=y, cmap="bwr")
plt.scatter(X_new[:, 0], X_new[:, 1], c='white', edgecolor="k")

plt.show()
```

```python
# initialisation du model √† k = 3
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X, y)
```

```python
# effectuer une prediction
predictions = model.predict(X_new)

plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.4)
plt.scatter(X_new[:, 0], X_new[:, 1], c=predictions, edgecolor="k")

plt.show()
```

```python
plt.figure(figsize=(12, 4))
    
    #  Train Loss
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="bwr")
plt.scatter(X_new[:, 0], X_new[:, 1], c='green', edgecolor="k")

    #  Train Loss
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.4)
plt.scatter(X_new[:, 0], X_new[:, 1], c=predictions, edgecolor="k")

plt.show()

```

```python
plt.figure(figsize=(12, 5))  # Slightly taller to accommodate colorbar

# First subplot - Original data with new points
plt.subplot(1, 2, 1)
scatter1 = plt.scatter(X[:, 0], X[:, 1], c=y, cmap="bwr", label='Original Data')
scatter2 = plt.scatter(X_new[:, 0], X_new[:, 1], c='green', edgecolor="k", 
                      label='New Points', s=100)
plt.title("Original Data with New Points")
plt.legend()

# Second subplot - Predictions on new points
plt.subplot(1, 2, 2)
scatter3 = plt.scatter(X[:, 0], X[:, 1], c=y, cmap="bwr", alpha=0.4, 
                      label='Original Data')
scatter4 = plt.scatter(X_new[:, 0], X_new[:, 1], c=predictions, cmap="bwr", 
                      edgecolor="k", s=100, label='Predictions')
plt.title("Predictions on New Points")
plt.legend()



plt.subplots_adjust(top=0.85, bottom=0.15)  # Adjust spacing
plt.show()
```
