Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

plt.style.use('dark_background')  # üï∂Ô∏è Dark mode
```

```python
np.random.seed(0)

m = 50
X = np.random.randn(m).reshape(-1, 1)
y = 1.5 * X + 3 + np.random.randn(m).reshape(-1, 1)

plt.scatter(X, y)
plt.show()
```

Utilisation du modele LIneraRegression de scikit-learn

```python
model = LinearRegression()
model.fit(X, y)

new_data = np.linspace(-3, 3).reshape(-1, 1)
predictions = model.predict(new_data)

plt.scatter(X, y)
plt.plot(new_data, predictions, c='red')
plt.show()
```
