Clustering - Mod√®le KMeans

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

plt.style.use('dark_background')  # üï∂Ô∏è Dark mode
```

```python
# creation d'un jeu de donees avec un nombre de cluster al√©atoire
k = np.random.choice([2, 3, 4, 5, 6, 7, 8])

X, _ = make_blobs(n_samples=200, centers=k, n_features=2, random_state=0, cluster_std=0.5)
```

```python
# entrainement du model KMeans
k_range = range(1, 10)
k_inertias = []

for k in k_range:
    model = KMeans(n_clusters=k).fit(X)
    k_inertias.append(model.inertia_)

plt.plot(k_range, k_inertias, lw=3)
plt.show()
```

```python
# creation d'un models avec K relatif au coude ci haut
model = KMeans(n_clusters=6)
model.fit(X)

predictions = model.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=predictions, edgecolor="k")

plt.show()
```
